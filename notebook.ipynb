```python
import pandas as pd

# Read the csv file
df = pd.read_csv('./tabelas/available_data.tsv', sep='\t')
df.head()
```

```python
# Read the products
df_products = pd.read_csv('./tabelas/df_dummies_products.csv', sep=';')
df_products.head()

# Split the column 'mes' in 'ano' and 'mes'
df_products['ano'] = df_products['mes'].apply(lambda x: int(x.split('-')[0]))
df_products['mes'] = df_products['mes'].apply(lambda x: int(x.split('-')[1]))
```

```python
# get all rows which tx_est_id is 26626
df_26626 = df_products.loc[df_products['tx_est_id'] == 26626]
df_26626.head()
```

```python
# Create new empty dataframe to merge the products to the available data
df_new = pd.DataFrame(columns=df.columns.tolist() + df_products.columns.tolist())

# get unique values from 'rec_est_id' from df
unique_ids_from_df = df['rec_est_id'].unique()

concatenated_rows = []

count_ids = 0
for id in unique_ids_from_df:
    # get all rows from df which rec_est_id is equal to id
    df_id = df.loc[df['rec_est_id'] == id]

    # check if id is in df_products
    if id in df_products['tx_est_id'].unique():
        # get all rows from df_products which tx_est_id is equal to id
        df_products_by_id = df_products.loc[df_products['tx_est_id'] == id].sort_values(by=['ano', 'mes'])
        
        print(df_products_by_id.head())
        print(df_id.head())
              
        # iterate over all rows from df_id
        for _, row1 in df_id.iterrows():
            month = row1['rec_month_part']
            year = row1['rec_year_part']

            # print whole row that matches the condition
            filtered_rows = df_products_by_id[
                (year == df_products_by_id['ano']) & 
                (month == df_products_by_id['mes'])
                ]
            print(filtered_rows)
                    
            # Concatenar as linhas correspondentes da tabela 1 e da tabela 2
            for _, row2 in filtered_rows.iterrows():
                concatenated_row = {**row1, **row2}  # Concatenar os dicionários das linhas
                concatenated_rows.append(concatenated_row)
            
        count_ids += 1
    print(f'{count_ids} of {len(unique_ids_from_df)}')
    if count_ids == 3:
        break
    print(concatenated_rows)
```
